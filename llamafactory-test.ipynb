```python
import torch
from llamafactory.train.tuner import run_exp

# 设置设备（如果有GPU）
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
```

```python
def train_model():
    """使用Python API训练模型"""
    
    # 定义训练参数
    train_args = {
        # 模型配置
        "model_name_or_path": "../modelscope/Qwen/Qwen2___5-Math-1___5B",  # 基础模型
        "output_dir": "../outputs",  # 输出目录
        
        # 数据配置
        "dataset": "train",  # 使用自定义数据集，这里的名称应与dataset_info.json中的键名匹配
        "dataset_dir": "./data",  # 数据目录路径
        "template": "default",  # 模板类型
        
        # 训练配置
        "finetuning_type": "lora",  # 微调类型：lora, full, freeze
        "num_train_epochs": 3.0,
        "learning_rate": 1e-4,
        "per_device_train_batch_size": 2,
        "gradient_accumulation_steps": 4,
        
        # LoRA配置（如果使用LoRA）
        "lora_rank": 16,
        "lora_alpha": 32,
        "lora_dropout": 0.1,
        "lora_target": "q_proj,v_proj",  # 目标模块
        
        # 其他配置
        "logging_steps": 10,
        "save_steps": 500,
        "fp16": True,  # 如果GPU支持混合精度
        "overwrite_output_dir": True,
    }
    
    # 开始训练
    run_exp(train_args)
    
    print("训练完成！模型保存在:", train_args["output_dir"])

# 执行训练
train_model()
```
